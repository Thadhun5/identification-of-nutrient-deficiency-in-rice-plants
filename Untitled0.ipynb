```python
x_train, x_val, y_train, y_val = train_test_split(x_train_parts, y_train_parts, test_size=0.1, random_state=42)
encoded_labels_train = to_categorical(y_train, num_classes=3)
encoded_labels_val = to_categorical(y_val, num_classes=3)
```

```python
from flask import Flask, render_template, request
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import base64

app = Flask(__name__)

# Load your pre-trained model
loaded_model = load_model('/content/Trained data/deciese')  # Replace with the correct path

# Define label mapping
label_mapping = {
    0: 'NITROGEN',
    1: 'PHOSPHORUS',
    2: 'POTASSIUM'
}

@app.route('/', methods=['GET', 'POST'])
def predict():
    prediction_result = None
    selected_image = None

    if request.method == 'POST':
        file = request.files['file']
        image = cv2.imdecode(np.fromstring(file.read(), np.uint8), cv2.IMREAD_COLOR)

        # Resize the image directly
        image = cv2.resize(image, (256, 1280))

        # Divide the image into parts using numpy operations
        image_parts = np.array_split(image, 5)

        # Convert to float and normalize pixel values
        image_parts = np.array(image_parts, dtype='float32') / 255

        # Reshape the image parts array to match the input shape expected by the model
        image_parts = image_parts.reshape((-1, 256, 256, 3))

        # Make predictions
        predictions = loaded_model.predict(image_parts)

        # Get the class with the highest probability for each part
        predicted_classes = np.argmax(predictions, axis=1)
        predicted_labels = [label_mapping[p] for p in predicted_classes]

        # Calculate the combined probability and majority class
        average_probability = np.mean(np.max(predictions, axis=1))
        majority_class = np.argmax(np.bincount(predicted_classes))
        prediction_result = f"Combined Predicted Label (Majority Class): {label_mapping[majority_class]}, Combined Predicted Probability (Average): {average_probability:.4f}"

        # Convert the selected image to base64 encoding
        _, buffer = cv2.imencode('.jpg', image)
        image_encoded = base64.b64encode(buffer).decode('utf-8')
        selected_image = f"data:image/jpeg;base64,{image_encoded}"

    return render_template('/content/HTML file/index.html', prediction_result=prediction_result, selected_image=selected_image)

if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=8123)
```
